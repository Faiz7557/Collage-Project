```python
import pandas as pd
import numpy as np

# 1. Load ketiga dataset
df1 = pd.read_csv("C:/Uner/Semester 5/Data Mining II/Coolyeah/UAS/dataset/kuhap.csv")
df2 = pd.read_csv("C:/Uner/Semester 5/Data Mining II/Coolyeah/UAS/dataset/ruu_kuhap.csv")
df3 = pd.read_csv("C:/Uner/Semester 5/Data Mining II/Coolyeah/UAS/dataset/ruu_tweet.csv")

# 2. Gabungkan menjadi satu dataset utuh
combined_df = pd.concat([df1, df2, df3], ignore_index=True)

# 3. Bersihkan data duplikat berdasarkan ID tweet
combined_df = combined_df.drop_duplicates(subset='id_str')

# 4. Konversi kolom waktu ke format datetime agar bisa diurutkan
# (Pandas biasanya otomatis mendeteksi formatnya)
combined_df['created_at'] = pd.to_datetime(combined_df['created_at'])

# 5. Urutkan dari yang terlama ke terbaru
combined_df = combined_df.sort_values(by='created_at', ascending=True)
combined_df = combined_df.reset_index(drop=True)

# 6. Lakukan Sampling (Systematic Sampling) untuk mendapatkan 2000 baris
# Metode ini mengambil data dengan interval tetap sehingga sebarannya merata dari awal hingga akhir waktu
target_rows = 2000
total_rows = len(combined_df)

if total_rows > target_rows:
    # Membuat array indeks dengan jarak yang sama (linspace)
    indices = np.linspace(0, total_rows - 1, target_rows, dtype=int)
    sampled_df = combined_df.iloc[indices].copy()
else:
    # Jika data kurang dari 2000, ambil semuanya
    sampled_df = combined_df.copy()

# 7. Simpan hasil ke file CSV baru
sampled_df.to_csv('data_kuhap.csv', index=False)

print(f"Proses selesai.")
print(f"Total data sebelum sampling: {total_rows}")
print(f"Total data setelah sampling: {len(sampled_df)}")
```
