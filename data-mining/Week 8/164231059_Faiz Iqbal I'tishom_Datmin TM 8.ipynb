Sebuah ANN sederhana digunakan untuk memprediksi nilai skor suatu sistem berdasarkan 3 fitur masukan. Sistem ini memiliki arsitektur sebagai berikut:

    1. Input layer: 3 neuron (X = np.array([1.0, 2.0, -1.0]))

    2. Hidden layer: 4 neuron dengan aktivasi ReLU

    3. Output layer: 1 neuron dengan aktivasi linear (tanpa aktivasi)

    4. Target Output (y_true) = 3.5

    5. Learning rate = 0.1

Bobot dan bias diinisialisasi secara acak (jangan lupa gunakan seed). Target output berupa nilai kontinu (float). Fungsi loss yang digunakan adalah Mean Squared Error (MSE) dan dihitung diluar fungsi forward_pass().

```python
import numpy as np
```

```python
def relu(x):
    return np.maximum(0, x)

def mean_squared_error(y_pred, y_true):
    return 0.5 * (y_pred - y_true) ** 2
```

## Algoritma ANN

```python
# Seed agar hasil random bisa direproduksi
np.random.seed(27229)

# Input data
X = np.array([1.0, 2.0, -1.0])  # data shape: (3,)
y_true = 3.5
learning_rate = 0.1

# -------------------------
# Inisialisasi bobot & bias
# -------------------------

# Input (3) -> dengan Hidden Layer (4)
W1 = np.random.randn(4, 3)  # shape: (4, 3)
b1 = np.random.randn(4)     # shape: (4,)

# Hidden Layer (4) -> Output (1)
W2 = np.random.randn(1, 4)  # shape: (1, 4)
b2 = np.random.randn(1)     # shape: (1,)

# -------------------------
# Fungsi aktivasi ReLU dan turunannya
# -------------------------

def relu(z):
    return np.maximum(0, z)

def relu_derivative(z):
    return (z > 0).astype(float)

# -------------------------
# Fungsi Forward pass
# -------------------------

def forward_pass(X, W1, b1, W2, b2):
    # Layer 1: Input -> Hidden
    z1 = W1 @ X + b1        # shape: (4,)
    a1 = relu(z1)           # shape: (4,)

    # Layer 2: Hidden -> Output
    z2 = W2 @ a1 + b2       # shape: (1,)
    y_pred = z2[0]          # scalar output

    #Output Fungsi
    return z1, a1, y_pred

# -------------------------
# Training step (1 epoch)
# -------------------------

# Forward pass
z1, a1, y_pred = forward_pass(X, W1, b1, W2, b2)

# Hitung loss: Mean Squared Error
loss = 0.5 * (y_true - y_pred)**2
print("Loss Tanpa Backpropagation:", loss)

# -------------------------
# Backpropagation
# -------------------------

# Gradient dari loss terhadap y_pred
dL_dy_pred = y_pred - y_true  # scalar

# Gradien untuk W2 dan b2 (output layer)
dL_dW2 = dL_dy_pred * a1.reshape(1, -1)  # shape: (1, 4)
dL_db2 = dL_dy_pred                    # shape: (1,)

# Gradien untuk hidden layer
dL_da1 = dL_dy_pred * W2.reshape(-1)   # shape: (4,)
dL_dz1 = dL_da1 * relu_derivative(z1)   # shape: (4,)

# Gradien untuk W1 dan b1
dL_dW1 = dL_dz1.reshape(-1, 1) @ X.reshape(1, -1)  # shape: (4, 3)
dL_db1 = dL_dz1                                    # shape: (4,)

# -------------------------
# Update bobot & bias
# -------------------------

W2 -= learning_rate * dL_dW2
b2 -= learning_rate * dL_db2
W1 -= learning_rate * dL_dW1
b1 -= learning_rate * dL_db1

# -------------------------
# Cek hasil setelah update
# -------------------------
_, _, y_pred_updated = forward_pass(X, W1, b1, W2, b2)
loss_updated = 0.5 * (y_true - y_pred_updated)**2
print("Loss Dengan Backpropagation:", loss_updated)
```

## Dengan Epoch

```python
# Seed agar reproducible
np.random.seed(27229)

# Input dan targeted value
X = np.array([1.0, 2.0, -1.0])
y_true = 3.5
learning_rate = 0.1
epochs = 100

# Inisialisasi bobot dan bias pada fungsi
W1 = np.random.randn(4, 3)
b1 = np.random.randn(4)
W2 = np.random.randn(1, 4)
b2 = np.random.randn(1)

# ReLU dan turunannya
def relu(z):
    return np.maximum(0, z)

def relu_derivative(z):
    return (z > 0).astype(float)

# Forward pass
def forward_pass(X, W1, b1, W2, b2):
    z1 = W1 @ X + b1
    a1 = relu(z1)
    z2 = W2 @ a1 + b2
    y_pred = z2[0]
    return z1, a1, y_pred

# Training loop
for epoch in range(epochs):
    # Forward pass
    z1, a1, y_pred = forward_pass(X, W1, b1, W2, b2)

    # Hitung loss
    loss = 0.5 * (y_true - y_pred) ** 2

    # Backpropagation
    dL_dy_pred = y_pred - y_true

    # Gradien output layer
    dL_dW2 = dL_dy_pred * a1.reshape(1, -1)
    dL_db2 = dL_dy_pred

    # Gradien hidden layer
    dL_da1 = dL_dy_pred * W2.reshape(-1)
    dL_dz1 = dL_da1 * relu_derivative(z1)

    dL_dW1 = dL_dz1.reshape(-1, 1) @ X.reshape(1, -1)
    dL_db1 = dL_dz1

    # Update parameter
    W2 -= learning_rate * dL_dW2
    b2 -= learning_rate * dL_db2
    W1 -= learning_rate * dL_dW1
    b1 -= learning_rate * dL_db1

    # Print setiap 5 epoch
    if (epoch + 1) % 1 == 0 or epoch == 0:
        print(f"Epoch {epoch + 1:3d} | Loss: {loss:.4f} | y_pred: {y_pred:.4f}")
```
