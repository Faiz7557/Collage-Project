### Import Library

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report
```

### Import Dataset

```python
df = pd.read_excel("C:/Uner/Semester 4/Data Mining I/Week 7/data jeruk.xlsx")
```

```python
x = df[['R', 'G', 'B']].values
y = df['RASA'].values
```

### Labeling

```python
le = LabelEncoder()
y = le.fit_transform(y)
```

### Split Data

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.25)
```

### Perhitungan KNN secara manual

```python
def euclidean_distance(a, b):
    return np.sqrt(np.sum((a - b) ** 2))
```

```python
def knn_manual(X_train, y_train, X_test, k=5):
    y_pred = []
    for test_point in X_test:
        distances = [euclidean_distance(test_point, train_point) for train_point in X_train]
        k_indices = np.argsort(distances)[:k]
        k_labels = [y_train[i] for i in k_indices]
        predicted_label = max(set(k_labels), key=k_labels.count)
        y_pred.append(predicted_label)
    return np.array(y_pred)
```

```python
y_pred_manual = knn_manual(X_train, y_train, X_test, k=5)
```

```python
print("Hasil Perhitungan KNN Manual:")
print(classification_report(y_test, y_pred_manual, target_names=le.classes_))
```

### Perhitungan nilai NIM

```python
def euclidean_distance(a, b):
    return np.sqrt(np.sum((a - b) ** 2))

def knn_manual(X_train, y_train, X_test, k=4):
    y_pred = []
    for test_point in X_test:
        distances = [euclidean_distance(test_point, train_point) for train_point in X_train]
        k_indices = np.argsort(distances)[:k]
        k_labels = [y_train[i] for i in k_indices]
        predicted_label = max(set(k_labels), key=k_labels.count)
        y_pred.append(predicted_label)
    return np.array(y_pred)
```

```python
r, g, b = 100, 100, 59
nim = np.array([[r, g, b]])
```

```python
y_pred_nim = knn_manual(X_train, y_train, nim, k=5)

print("Prediksi untuk RGB (100, 100, 59):", y_pred_nim[0])
```
