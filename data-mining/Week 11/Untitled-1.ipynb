```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt

# Load shapefile negara
world = gpd.read_file('C:/Uner/Lomba/Nacoesta/ne_10m_admin_0_countries.shp')

# Load data CSV kamu (sudah kamu upload)
data = pd.read_csv("C:/Uner/Lomba/Nacoesta/data_spasial.csv")

# 3. Gabungkan berdasarkan nama negara
# Shapefile: kolom 'ADMIN', CSV: kolom 'Entity'
world_merged = world.merge(data, left_on="ADMIN", right_on="Entity", how="left")

# 4. Visualisasikan peta berdasarkan GDP per kapita
fig, ax = plt.subplots(1, 1, figsize=(15, 10))
world_merged.plot(
    column='GDP per capita, PPP (constant 2021 international $)',
    cmap='YlGnBu',
    linewidth=0.8,
    ax=ax,
    edgecolor='0.8',
    legend=True
)

ax.set_title("Peta GDP per Kapita Tahun 2022 (PPP, $ 2021)", fontsize=16)
ax.axis('off')
plt.show()
```
