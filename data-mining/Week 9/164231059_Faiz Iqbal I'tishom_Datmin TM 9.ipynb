### Inisialisasi input, trageted value, bobot, dan bias pada fungsi


```python
X = np.array([1.0, 2.0, -1.0])
y_true = 3.5
W1 = np.random.randn(4, 3)
b1 = np.random.randn(4)
W2 = np.random.randn(1, 4)
b2 = np.random.randn(1)
```

### Inisialisasi fungsi aktivasi

```python
def relu(x):
    return np.maximum(0, x)

def mean_squared_error(y_pred, y_true):
    return 0.5 * (y_pred - y_true) ** 2
```

### Forward Pass

```python
def forward_pass(X, W1, b1, W2, b2):
    hidden_input = np.dot(W1, X) + b1
    hidden_output = relu(hidden_input)
    final_input = np.dot(W2, hidden_output) + b2
    y_pred = final_input[0]
    return y_pred, hidden_output, hidden_input, final_input
```

### Backward Pass

```python
def backward_pass(X, y_true, W1, b1, W2, b2, hidden_output, hidden_input, output_input, y_pred):
    dL_dy_pred = y_pred - y_true
    dW2 = dL_dy_pred * hidden_output.reshape(1, -1)
    db2 = dL_dy_pred

    dL_da1 = dL_dy_pred * W2.reshape(-1)
    dL_dz1 = dL_da1 * relu_derivative(hidden_input)

    dW1 = dL_dz1.reshape(-1, 1) @ X.reshape(1, -1)
    db1 = dL_dz1
    return dW2, db2, dW1, db1
```

## Backpropagation

### Inisialisasi seed, learning rate, dan epoch

```python
# Seed agar reproducible
np.random.seed(27229)
# Set format tampilan array agar lebih rapi
np.set_printoptions(precision=4, suppress=True)
# Input dan targeted value
learning_rate = 0.1
epochs = 3
```

### Inisialisasi Fungsi Relu dan Turunannya

```python
def relu(x):
    return np.maximum(0, x)
def relu_derivative(x):
    return (x > 0).astype(float)
```

### Forward Pass dan Backpropagation

```python
def train_regression_3epoch():
    np.random.seed(27229)
    X = np.array([1.0, 2.0, -1.0])
    y_true = 3.5
    W1 = np.random.randn(4, 3)
    b1 = np.random.randn(4)
    W2 = np.random.randn(1, 4)
    b2 = np.random.randn(1)
    learning_rate = 0.1

    for epoch in range(epochs):
        print(f"\n{'='*25}\nEpoch {epoch + 1}\n{'='*25}")

        # Forward pass
        y_pred, hidden_output, hidden_input, final_input = forward_pass(X, W1, b1, W2, b2)
        loss = 0.5 * (y_true - y_pred) ** 2

        print("\n>> INPUT")
        print(f"X          : {X}")
        print(f"Target y   : {y_true:.4f}")

        print("\n>> FORWARD PASS")
        print(f"W1         :\n{W1}")
        print("\n")
        print(f"b1         : {b1}")
        print("\n")
        print(f"z1 (Hidden Input)  : {hidden_input}")
        print("\n")
        print(f"a1 (Hidden Output) : {hidden_output}")
        print("\n")
        print(f"W2         :\n{W2}")
        print("\n")
        print(f"b2         : {b2}")
        print("\n")
        print(f"z2 (Final Input)   : {final_input}")
        print("\n")
        print(f"y_pred     : {y_pred:.4f}")
        print("\n")
        print(f"Loss       : {loss:.6f}")

        # Backward pass
        dW2, db2, dW1, db1 = backward_pass(X, y_true, W1, b1, W2, b2, hidden_output, hidden_input, final_input, y_pred)

        print("\n>> BACKWARD PASS (Gradients)")
        print(f"dW2        :\n{dW2}")
        print("\n")
        print(f"db2        : {db2:.6f}")
        print("\n")
        print(f"dW1        :\n{dW1}")
        print("\n")
        print(f"db1        : {db1}")

        # Update
        W2 -= learning_rate * dW2
        b2 -= learning_rate * db2
        W1 -= learning_rate * dW1
        b1 -= learning_rate * db1

        print("\n>> UPDATED PARAMETERS")
        print(f"Updated W1 :\n{W1}")
        print("\n")
        print(f"Updated b1 : {b1}")
        print("\n")
        print(f"Updated W2 :\n{W2}")
        print("\n")
        print(f"Updated b2 : {b2}")
        print("\n")
        print(f"y_pred : {y_pred:.4f} | Loss : {loss:.6f} | Akurasi: {1 - abs(y_pred - y_true):.6f} | ")

# Jalankan training
train_regression_3epoch()
```
